```python
from datetime import datetime
import pandas as pd
import pytz
```

```python
# file_path = "C:\\Users\\carol\\OneDrive\\Documentos\\Dados Sensor Conatus\\Teste23-07.csv"
file_path = "C:\\Users\\carol\\OneDrive\\Documentos\\Dados Sensor Conatus\\Turbidimetro\\Turb_27-08.csv"

df = pd.read_csv(file_path)
df
```

```python
# Defina o fuso horário desejado
timezone = pytz.timezone('America/Sao_Paulo')  # Exemplo de fuso horário para São Paulo

#outputfile = "C:\\Users\\carol\\OneDrive\\Documentos\\Dados Sensor Conatus\\DadosSensorConatus_23-07.csv"
outputfile = "C:\\Users\\carol\\OneDrive\\Documentos\\Dados Sensor Conatus\\Turbidimetro\\DadosTurbConatus_27-08.csv"


def convert_timestamp(timestamp_ms):
    # Converta o timestamp de milissegundos para segundos
    timestamp_s = timestamp_ms / 1000.0
    # Crie um objeto datetime a partir do timestamp em segundos
    dt_utc = datetime.utcfromtimestamp(timestamp_s)
    # Converta o datetime UTC para o fuso horário desejado
    dt_local = dt_utc.replace(tzinfo=pytz.utc).astimezone(timezone)
    return dt_local

# Aplique a função a todas as linhas da coluna 'timestamp_ms'
df['local_datetime'] = df['timestamp'].apply(convert_timestamp)

df.drop(columns='datetime', inplace=True)

df.to_csv(outputfile, index=False)

print(df)
```
